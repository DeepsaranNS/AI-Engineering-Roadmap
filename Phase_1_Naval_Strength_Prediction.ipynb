```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 1. Geopolitical Naval Fleet Dataset (Hypothetical Strength Indexes)
data = {
    "Country": ["USA", "China", "India", "Russia", "UK", "France", "Japan", "South Korea"],
    "Submarines": [68, 79, 16, 58, 10, 9, 22, 22],
    "Destroyers": [92, 50, 12, 14, 6, 7, 36, 13],
    "Carriers": [11, 2, 2, 1, 2, 1, 4, 0],
    "Strength_Index": [98.5, 85.2, 55.4, 70.1, 45.0, 42.5, 60.3, 48.2] # The target we want to predict
}
fleet_df = pd.DataFrame(data)
print(fleet_df)
# 2. Separate Features (X) and Target (y)
X = fleet_df[['Submarines', 'Destroyers', 'Carriers']]
y = fleet_df['Strength_Index']
print(X)
print(y)
# 3. The Golden Rule: Split the data (75% for training, 25% locked in the vault for testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Train the Classical ML Model
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Make Predictions on the Test Set
predictions = model.predict(X_test)

print("--- MODEL PREDICTIONS vs ACTUAL STRENGTH ---")
for i, country_index in enumerate(X_test.index):
    country_name = fleet_df.loc[country_index, 'Country']
    actual = y_test.iloc[i]
    predicted = predictions[i]
    print(f"{country_name}: Predicted {predicted:.1f} | Actual {actual:.1f}")
```

```python
# The model stores its calculated weights in an attribute called 'coef_'
weights = model.coef_

print("--- THE MODEL'S MATHEMATICAL WEIGHTS ---")
# Using zip to iterate through both the feature names and their weights simultaneously
for ship_type, weight in zip(X.columns, weights):
    print(f"1 {ship_type} = {weight:.2f} Strength Points")
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# 1. Calculate the standard Mean Squared Error (MSE)
mse = mean_squared_error(y_test, predictions)

# 2. Take the square root to get RMSE (using numpy, the math library pandas is built on)
rmse = np.sqrt(mse)

print(f"The model's RMSE is: {rmse:.2f} Strength Points")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score

# 1. Add a Classification Target to our existing dataset
# 1 = "Strategic Power Projection" (Blue-Water capable)
# 0 = "Regional Defense"
# We assign these categories to our 8 nations: USA, China, India, Russia, UK, France, Japan, South Korea
fleet_df['Power_Projection'] = [1, 1, 1, 1, 1, 1, 0, 0]

# 2. Separate Features (X) and the new Target (y)
X = fleet_df[['Submarines', 'Destroyers', 'Carriers']]
y_class = fleet_df['Power_Projection']

# 3. Train/Test Split (Locking 25% away in the vault again)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_class, test_size=0.25, random_state=42)

# 4. Train the Logistic Regression Classifier
classifier = LogisticRegression()
classifier.fit(X_train_c, y_train_c)

# 5. Give it the Final Exam
class_predictions = classifier.predict(X_test_c)

# 6. Grade the Model using Classification Metrics
accuracy = accuracy_score(y_test_c, class_predictions)
precision = precision_score(y_test_c, class_predictions, zero_division=0)
recall = recall_score(y_test_c, class_predictions, zero_division=0)

print("--- CLASSIFICATION EXAM RESULTS ---")
print(f"Accuracy:  {accuracy * 100:.1f}%")
print(f"Precision: {precision * 100:.1f}%")
print(f"Recall:    {recall * 100:.1f}%")
```
